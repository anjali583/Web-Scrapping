```python
import requests
```

# Store the website

```python
website='https://ackodrive.com/collection/volkswagen-cars/'
```

```python
response = requests.get(website)
```

# To check if requests is accepted we use status code

```python
response.status_code
```

- 200 is the value that the request is accepted

# We need to create a soup object

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

# Results

```python
results = soup.find_all('div',{'data-testid':'listingcardesktop'})
```
